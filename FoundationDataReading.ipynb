```python
import numpy as np

a=np.array([[1,2],[3,4]])
b=a+3
print('a=%s\n'%a)
print('b=%s\n'%b)
c=np.dot(a,b)
d=np.multiply(a,b)
e=a*b
f=np.transpose(a)
g=np.linalg.inv(a)
print('c=%s\n'%c)
print('d=%s\n'%d)
print('e=%s\n'%e)
print('f=%s\n'%f)
print('g=%s'%g)
```

```python
import numpy as np
import matplotlib.pyplot as plt
plt.close()

num=50; low=0; mid=10; high=10
x=np.linspace(1,50,num)
y1=np.random.randint(low,high,num)
y2=np.random.randint(mid,3*high,num)
plt.plot(x,y1,'ko')
plt.plot(x,y2,'b--')
plt.legend(('plot1','plot2'))
plt.show()
```
