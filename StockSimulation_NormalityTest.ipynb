```python
import numpy as np
import matplotlib.pyplot as plt

N=252 ; S=np.zeros([N,1]); S[0]=100
vol=0.10; mu=0.15; T=1; dt=T/N
t=np.linspace(0,T,N)
e=np.random.normal(0,1,N)

for i in range(N-1) :
  S[i+1,0]=S[i,0]*np.exp((mu-0.5*vol**2)*dt+vol*e[i]*np.sqrt(dt))

plt.plot(t[:],S[:],'b-',linewidth=0.3)
plt.xlabel('Time')
plt.ylabel('Stock Price')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

N=252*3; S=np.zeros([N,1]); S[0]=100
vol=0.15; mu=0.08; T=3; dt=T/N
t=np.linspace(0,T,N)

plt.xlabel('Time')
plt.ylabel('Stock Price')

iter_num=10000
S_T=np.zeros([iter_num,1])

for k in range(0,iter_num):
  e=np.random.normal(0,1,N)
  for i in range(N-1):
    S[i+1]=S[i]*np.exp((mu-0.5*vol**2)*dt+vol*e[i]*np.sqrt(dt))
  S_T[k]=S[-1]
  plt.plot(t[:],S[:],'b-',linewidth=0.3)
plt.show()
```

```python
plt.hist(S_T,bins=30)
plt.xlabel('S_T')
plt.ylabel('Count')
plt.title('Histogram')
plt.show()

ln_ST=np.log(S_T)
plt.hist(ln_ST,bins=30)
plt.xlabel('ln_ST')
plt.ylabel('Count')
plt.title('Histogram')
plt.show()

print('ln(S_T) mean : %f' %np.mean(ln_ST))
print('ln(S_T) std: %f'%np.std(ln_ST))

def jb_test(x):
  x=np.array(x)
  n=x.size
  m=x.mean()
  v=np.mean((x-m)**2)
  s=np.mean((x-m)**3)/(v**1.5) #skewness
  k=np.mean((x-m)**4)/(v**2) #kurtosis
  jb=n/6 * (s**2+((k-3)**2)/4)
  #p-value(chi-square df=2)
  from scipy.stats import chi2
  p=1-chi2.cdf(jb,df=2)
  return jb,p,s,k

jb,p,s,k=jb_test(S_T)
print(f'JB={jb:.3f},p={p:.3g}, S={s:.3f}, K={k:.3f}')

jb,p,s,k = jb_test(ln_ST)
print(f'JB={jb:.3f},p={p:.3g}, S={s:.3f}, K={k:.3f}')
```
